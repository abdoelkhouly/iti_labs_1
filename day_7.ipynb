Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(0, 10, 1)
fav_food = {"Pizza": 10, "Burger": 8, "Sushi": 6, "Pasta": 9}

fig, ax = plt.subplots(2, 2, figsize=(10, 5))

ax[0, 0].plot(X, X/2)
ax[0, 0].set_title("X vs X/2")

ax[0, 1].scatter(np.random.rand(10), np.random.rand(10))
ax[0, 1].set_title("Random Scatter")

ax[1, 0].bar(fav_food.keys(), fav_food.values())
ax[1, 0].set_title("Favourite Foods")

rand_data = np.random.randn(1000)
ax[1, 1].hist(rand_data, bins=30)
ax[1, 1].set_title("Normal Distribution Histogram")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming car_sales DataFrame is already loaded from a previous cell execution
# If not, you would need to load it here:
# car_sales = pd.read_csv("/content/car-sales.csv")

# Convert 'Price' to numeric, create 'Total Sales' and 'Sale Date' columns
car_sales["Price"] = car_sales["Price"].str.replace('[\$,]', '', regex=True).astype(float)
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales["Sale Date"] = pd.to_datetime(pd.date_range(start='1/1/2020', periods=len(car_sales)))


car_sales.plot(x="Sale Date", y="Total Sales", kind="line", figsize=(10, 6))

plt.title("Total Sales Over Time")
plt.xlabel("Sale Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)

car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter", figsize=(8,6))

plt.title("Price vs Odometer (KM)")
plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python

X = np.random.rand(10, 4)

df = pd.DataFrame(X, columns=['a', 'b', 'c', 'd'])

df.plot(kind="bar", figsize=(10,6))

plt.title("Bar Graph of Random Numbers")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend(title="Columns")
plt.show()
```

```python

car_sales.groupby("Make")["Odometer (KM)"].mean().plot(kind="bar", figsize=(8,5))

plt.title("Average Odometer (KM) per Car Make")
plt.xlabel("Car Make")
plt.ylabel("Average Odometer (KM)")
plt.show()
```

```python

car_sales["Odometer (KM)"].plot(kind="hist", bins=30, figsize=(8,5))

plt.title("Distribution of Odometer (KM)")
plt.xlabel("Odometer (KM)")
plt.ylabel("Frequency")
plt.show()
```
