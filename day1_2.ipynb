```python
# 1) Odd or Even
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is Even")  # باقي القسمة على 2 = 0 → Even
else:
    print(f"{num} is Odd")   # Otherwise → Odd

print("\n---\n")

# 2) Pass or Fail
score = int(input("Enter student score (0-100): "))
if score >= 50:
    print("Pass")
else:
    print("Fail")

print("\n---\n")

# 3) Largest of 3 numbers
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))
largest = max(a, b, c)  # max() ترجع أكبر قيمة
print("The largest number is:", largest)

print("\n---\n")

# 4) Counting with for (1 to 10)
print("Counting from 1 to 10:")
for i in range(1, 11):
    print(i, end=" ")  # end=" " يخلي الأرقام في نفس السطر
print()

print("\n---\n")

# 5) Grade Calculator
score = int(input("Enter score for grade calculation: "))
if 90 <= score <= 100:
    print("Grade: A")
elif 80 <= score <= 89:
    print("Grade: B")
elif 70 <= score <= 79:
    print("Grade: C")
elif 60 <= score <= 69:
    print("Grade: D")
elif score < 60:
    print("Grade: F")
else:
    print("Invalid score!")

print("\n---\n")

# 6) Multiplication Table
num = int(input("Enter a number for multiplication table: "))
print(f"Multiplication Table for {num}:")
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")

print("\n---\n")

# 7) Sum of Even Numbers (1 to 50)
sum_even = 0
for i in range(1, 51):
    if i % 2 == 0:
        sum_even += i
print("Sum of even numbers between 1 and 50 =", sum_even)

print("\n---\n")

# 8) Factorial with while loop
n = int(input("Enter a number to find factorial: "))
factorial = 1
counter = 1
while counter <= n:
    factorial *= counter
    counter += 1
print(f"Factorial of {n} is {factorial}")
```
